```python
# This documentation summarizes some key equations of KIDs
import numpy as np
from astropy import units as u
from astropy import constants as const
import matplotlib.pyplot as plt
def get_NEPphoton2(f,bw,Q):
    nep2 = 2*Q*const.h*f + Q**2/bw
    return nep2
```

```python
X_r300_oe4=np.loadtxt('/Users/bicep/Documents/DeviceDesign/MKID_SNF/slim_r300_oe0.4_loading.txt')
X_r1000_oe4=np.loadtxt('/Users/bicep/Documents/DeviceDesign/MKID_SNF/slim_r1000_oe0.4_loading.txt')
X_r300_oe2=np.loadtxt('/Users/bicep/Documents/DeviceDesign/MKID_SNF/slim_r300_oe0.2_loading.txt')
X_r300_oe4_space=np.loadtxt('/Users/bicep/Documents/DeviceDesign/MKID_SNF/slim_r300_oe0.4_spaceloading.txt')
X_r1000_oe4_space=np.loadtxt('/Users/bicep/Documents/DeviceDesign/MKID_SNF/slim_r1000_oe0.4_spaceloading.txt')
X_r300_oe2_space=np.loadtxt('/Users/bicep/Documents/DeviceDesign/MKID_SNF/slim_r300_oe0.2_spaceloading.txt')
f=X_r300_oe4[:,0]*u.GHz
bw=X_r300_oe4[:,1]*u.GHz
Qtot=X_r300_oe4[:,8]*u.pW
nep_ph2 = (get_NEPphoton2(f,bw,Qtot)).to(' W^2/Hz')
nep=nep_ph2**0.5
plt.plot(f,nep,color='b',label='R=300, oe=40%')
f=X_r300_oe4_space[:,0]*u.GHz
bw=X_r300_oe4_space[:,1]*u.GHz
Qtot=X_r300_oe4_space[:,8]*u.pW
nep_ph2 = (get_NEPphoton2(f,bw,Qtot)).to(' W^2/Hz')
nep=nep_ph2**0.5
plt.plot(f,nep,color='b',linestyle='--',label='R=300, oe=40%, no atmos')

f=X_r1000_oe4[:,0]*u.GHz
bw=X_r1000_oe4[:,1]*u.GHz
Qtot=X_r1000_oe4[:,8]*u.pW
nep_ph2 = (get_NEPphoton2(f,bw,Qtot)).to(' W^2/Hz')
nep=nep_ph2**0.5
plt.plot(f,nep,color='r',label='R=1000, oe=40%')
f=X_r1000_oe4_space[:,0]*u.GHz
bw=X_r1000_oe4_space[:,1]*u.GHz
Qtot=X_r1000_oe4_space[:,8]*u.pW
nep_ph2 = (get_NEPphoton2(f,bw,Qtot)).to(' W^2/Hz')
nep=nep_ph2**0.5
plt.plot(f,nep,color='r',linestyle='--',label='R=1000, oe=40%, no atmos')

f=X_r300_oe2[:,0]*u.GHz
bw=X_r300_oe2[:,1]*u.GHz
Qtot=X_r300_oe2[:,8]*u.pW
nep_ph2 = (get_NEPphoton2(f,bw,Qtot)).to(' W^2/Hz')
nep=nep_ph2**0.5
plt.plot(f,nep,color='k',label='R=300, oe=20%')
f=X_r300_oe2_space[:,0]*u.GHz
bw=X_r300_oe2_space[:,1]*u.GHz
Qtot=X_r300_oe2_space[:,8]*u.pW
nep_ph2 = (get_NEPphoton2(f,bw,Qtot)).to(' W^2/Hz')
nep=nep_ph2**0.5
plt.plot(f,nep,color='k',linestyle='--',label='R=300, oe=20%, no atmos')

plt.xlim(80,380)
plt.ylim(1e-20,1e-16)
plt.yscale('log')
plt.xlabel('GHz')
plt.ylabel('W/rtHz')
plt.title('photon noise NEP')
plt.grid()
plt.legend(loc='lower left', bbox_to_anchor=(1, 0.5))
```
